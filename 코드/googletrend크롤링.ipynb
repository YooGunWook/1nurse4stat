## 아래 과정은 가수 이름을 추출하기 위함이다

```python
singername = list()
for i in np.arange(len(singer)):
    singername.append(singer[i])
```

```python
temp = set(singername)
temp = list(temp)
```

```python
del temp[0]
```

```python
singer = list()
for i in np.arange(len(temp)):
    singer = singer + temp[i].split(",")
```

```python
import re
```

```python
m = re.compile(".+(?=\()")
```

```python
singername = list()
```

```python
for i in np.arange(len(singer)):
    t = m.search(singer[i])
    if t == None:
        singername.append(singer[i])
    else:
        singername.append(t.group())
    
    
```

```python
singer = list()
for i in np.arange(len(singername)):
    singer.append(singername[i].strip())
```

```python
temp = set(singer)
singer = list(temp)
```

## Google Trend 크롤링

```python
import time
```

```python
singertrenddf = pd.DataFrame()

for i in np.arange(len(singer)):
    pytrend = TrendReq(tz = 480,retries = 4, geo='KR', backoff_factor=0.05)
    
    keyword = singer[i]
     
    pytrend.build_payload(kw_list=[keyword], timeframe='2018-04-22 2020-04-25',   geo="KR")

    singertrendtemp = pytrend.interest_over_time()
    
    if str(singertrendtemp) != 'Empty DataFrame\nColumns: []\nIndex: []':

        singertrenddf = singertrenddf.append(singertrendtemp.iloc[:, 0])
    
```

```python
response = uncrawledsinger
```

```python
uncrawledartist = pd.DataFrame()
for i in np.arange(0,len(uncrawledsinger)):
    
    pytrend = TrendReq(tz = 480,retries = 4, geo='KR', backoff_factor=0.05)
     
    pytrend.build_payload(kw_list=[response[i]], timeframe='2019-06-01 2020-04-19',   geo="KR")

    singertrendtemp = pytrend.interest_over_time()
    
    if str(singertrendtemp) != 'Empty DataFrame\nColumns: []\nIndex: []':

        uncrawledartist = uncrawledartist.append(singertrendtemp.iloc[:, 0])
```

```python
uncrawled_name = uncrawledartist.index.tolist()

for i in range(0, len(uncrawled_name)):
    name = uncrawled_name[i]
    uncrawled_name[i] = name.replace(" ","")
```

```python
uncrawledartist.index = uncrawled_name
```

```python
singertrenddf = singertrenddf.append(uncrawledartist)
```
