```python
# 크롤링

for id in id_all :
    tf = 1
    
    soup = []
    dc = []
    date = []
    view = []
    recmd = []
    numb = []
    singer = []
    
    print(id[1])
    
    while tf:
        base_url = "https://gall.dcinside.com/mgallery/board/lists/"
        params = {'id': str(id[0]),'page':tf}
        headers = {'User-Agent' :'*' }
        
        resp = requests.get(base_url, params=params, headers=headers)
        soup = BeautifulSoup(resp.content, 'html.parser')
        
        try :
            contents = soup.find('tbody').find_all('tr',{'class':'ub-content us-post'})
            for i in contents:
                
                # 날짜 추출 
                date_tag = i.find('td', class_='gall_date')
                if date_tag :
                    date_dict = date_tag.attrs
                    if len(date_tag.attrs) is 2:
                        date.append(date_dict['title'])
                    else:
                        date.append(date_tag.text)
                        pass
                    
                # 조회 수 추출
                    views_tag = i.find('td', class_='gall_count')
                    view.append(views_tag.text)
                    
                # 추천 수 추출
                    recommend_tag = i.find('td', class_='gall_recommend')
                    recmd.append(recommend_tag.text)
                    
                # 게시물 number 추출
                    numb_tag = i.find('td', class_='gall_num')
                    numb.append(numb_tag.text)
                    
                # 가수명
                    singer.append(id[1])
        except :
            print("파인드올 에러 : " + id[1] ) 
            tf=-1

        tf = tf+1
        try :
            if (date[-1] < str('2018-04-01 00:00:01')) or (int(numb[-1]) < 50) : tf = False
        except :
            print("이프절 에러 : " + id[1] ) 
            tf=0
        
    try :
        data_frame = pd.DataFrame(dc)
        data_frame['date'] = date
        data_frame['views'] = view
        data_frame['recommend'] = recmd
        data_frame['number'] = numb
        data_frame['singer'] = singer
        
        data_frame.to_csv('C:/Users/User/Documents/google/R/desaip/' + id[1] + '.csv')
    except :
        print("데이터프레임 에러 : " + id[1] )
```
