```python
# 시작과 끝을 분리해준다.
week = df_youtube['주간'].str.split('-')
# 원래 데이터와 형태 맞춰주기 위해
# 그리고 유튜브 주기랑 가온차트 주기가 다르다. 
# 5를 빼주면 가온차트와 같아진다. 
start = []
end = []
for i in week:
    start.append(str(datetime.datetime.strptime(i[0],'%Y%m%d') - datetime.timedelta(days = 5))[0:10])
    end.append(str(datetime.datetime.strptime(i[1],'%Y%m%d') - datetime.timedelta(days = 5))[0:10])
```

```python
# 원래 데이터와 형태 맞춰주기 위해
# 그리고 유튜브 주기랑 가온차트 주기가 다르다. 
# 5를 빼주면 가온차트와 같아진다. 
start = []
end = []
for i in week:
    start.append(str(datetime.datetime.strptime(i[0],'%Y%m%d') - datetime.timedelta(days = 5))[0:10])
    end.append(str(datetime.datetime.strptime(i[1],'%Y%m%d') - datetime.timedelta(days = 5))[0:10])
```

```python
# 전처리한 값 추가
df_youtube['start'] = start
df_youtube['end'] = end
```

```python
df_youtube.head()
```

```python
# song과 날짜로 합칠거라서 song 전처리 해준다 
# 이유: 이름이 다른 사람들이 존재 ex) 조정석 == Jo Jung-suk
song = []
for i in df_youtube['song']:
    a = re.split('[(-)]+' ,i)[0]
    song.append(a.strip())
```

```python
df_youtube['song'] = song
```

```python
# 필요없는 값 제거
df_for_merge = df_youtube.drop(columns = ['주간','lastweek_rank','지속기간','변경','name'])
```

```python
# 이름 변경
df_you = df_for_merge.rename(columns = {'start':'st_day', 'end':'ed_day','song':'name'})
```

```python
df_you
```

```python
a = []
for i in df_you['st_day']:
    a.append(i.split('-'))
```

```python
year = []
day = []
month = []
for i in range(0,len(a)):
    years = int(a[i][0][2:4])
    months = int(a[i][1].replace('0',''))
    days = int(a[i][2])
    year.append(years)
    month.append(months)
    day.append(days)
```

```python
df_you['year'] = year
df_you['month'] = month
df_you['day'] = day
```

```python
df_you
```

```python
# 유튜브와 합치기
a = pd.merge(df,df_you,on=['st_day','ed_day','name'],how = 'left')
```

```python
# 필요 없는 값 제거
del a['id']
```

```python
# NA값 0으로 처리
a['순위'] = a['순위'].fillna(0)
a['조회수'] = a['조회수'].fillna(0)
```

```python
순위 = []
조회수 = []
# 조회수 전처리
for i in a['조회수']:
    if i == 0:
        조회수.append(0)
    elif i != 0:
        if 'M' in i:
            b = i.replace('M','00000')
            if '.' in b:
                b = b.replace('.','')
            조회수.append(int(b))
        elif 'K' in i:
            b = i.replace('K','0000')
            if '.' in b:
                b = b.replace('.','')
            조회수.append(int(b))

# 순위 전처리
for i in a['순위']:
    if i == 0:
        순위.append(i)
    elif i != 0:
        i = i.replace('위','')
        순위.append(int(i))
```

```python
a['순위'] = 순위
a['조회수'] = 조회수
a = a.rename(columns = {'순위':'you_rank'})
```
